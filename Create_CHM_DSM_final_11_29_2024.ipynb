```python
import os, glob
os.chdir(r'c:\Drive\githubs\pycrown')
import os
import glob
import numpy as np
import pandas as pd
import rasterio
import geopandas as gpd
import time
from pycrown.pycrown2 import PyCrown
from create_chm1 import chip_raster_bounding_box,ply2las,create_chm
from Pipelines_chm import create_dsm



def main():
    # Start timing the entire process
    # Start timing the entire process
    start_time = time.time()

    base_dir = r'C:\Users\durgj\OneDrive\Desktop\pycrown_hartford'
    output_dir = os.path.join(base_dir, 'chm_output')
    output_shapefile = os.path.join(base_dir, 'extent_bbox.shp')
    results = os.path.join(base_dir, 'results')
    
    os.makedirs(output_dir, exist_ok=True)  # Output folder
    os.makedirs(results, exist_ok=True)  # Result folder

    las_files = glob.glob(os.path.join(base_dir, r'Hartford_e_drive\*.las'))


    # Iterate over LAS files
    for las in las_files:  # Adjust as needed
        name = os.path.basename(las)
        base = name.split('.')[0]

        # Define file paths
        dem_path = os.path.join(output_dir, f'DEM\\{name.replace("las", "tif")}')
        dsm_path = os.path.join(output_dir, f'DSM\\{name.replace("las", "tif")}')
        resized_dsm_path = os.path.join(output_dir, f'DsM_Crp\\{name.replace("las", "tif")}')
        chm_path = os.path.join(output_dir, f'CHM\\{name.replace("las", "tif")}')
        crs_epsg = 6434

        # Create output directories if they do not exist
        os.makedirs(os.path.dirname(dsm_path), exist_ok=True)  # DSM folder
        os.makedirs(os.path.dirname(resized_dsm_path), exist_ok=True)  # Cropped DSM folder
        os.makedirs(os.path.dirname(chm_path), exist_ok=True)  # CHM folder

        # Create CHM and other necessary files if not already created
        create_dsm(las, dsm_path)  # Uncomment if needed
        create_chm(dsm_path, dem_path, resized_dsm_path, chm_path)  # Uncomment if needed
        file_check= os.path.join(r'C:\Users\durgj\OneDrive\Desktop\pycrown_hartford\results\CHM', os.path.basename(chm_path).split('.')[0]+'_chm_{}.tif'.format(3))
        if os.path.exists(file_check):
            print(f' {chm_path} already exists')
            continue
        else:
            print(f'working in {chm_path}')
        

            boxes = chip_raster_bounding_box(chm_path,2,2)

            for idx, bbox1 in enumerate(boxes):
                    print(f"Processing bbox {idx}: {bbox1}")
                    # Initialize PyCrown
                    PC = PyCrown(
                        chm_file=chm_path,
                        dtm_file=dem_path,
                        dsm_file=resized_dsm_path,
                        las_file=las,
                        outpath=results
                    )
                
                        # Process data
                    PC.clip_data_to_bbox(bbox=bbox1)
                    PC.filter_chm(6, ws_in_pixels=True, circular=False)
                    PC.tree_detection(PC.chm, ws=6, hmin=20.0,ws_in_pixels=True)
                    PC.crown_delineation(
                        algorithm='watershed_skimage',                      #####['dalponte_cython', 'dalponte_numba','dalponteCIRC_numba', 'watershed_skimage']
                        loc='top_cor',
                        th_tree=1.0, 
                        th_seed=2., 
                        th_crown=1.81, 
                        max_crown=60.
                    )
                    PC.correct_tree_tops()
                    PC.get_tree_height_elevation(loc='top')
                    PC.get_tree_height_elevation(loc='top_cor')
                    # PC.clip_trees_to_bbox(bbox=bbox1, loc='top_cor')
                    PC.screen_small_trees(hmin=20.0, loc='top_cor')
                    PC.crowns_to_polys_raster()
                    PC.crowns_to_polys_smooth(store_las=False,name=base, idx=idx)   
                    PC.quality_control()


                    # # # Export results
                    PC.export_raster(PC.chm, os.path.join(PC.outpath,'CHM', f'{base}_chm_{idx}.tif'), 'CHM')
                    # PC.export_tree_locations(loc='top', name=base, idx=idx)
                    PC.export_tree_locations(loc='top_cor', name=base, idx=idx)               
                    PC.export_tree_crowns(crowntype='crown_poly_raster',name=base, idx=idx)
                    PC.export_tree_crowns(crowntype='crown_poly_smooth',name=base, idx=idx)
                    print(f"Number of trees detected for {base} are : {len(PC.trees)}")

            # # Print total elapsed time
            print(f"Total time: {(time.time() - start_time)/60:.2f} minutes")

if __name__ == "__main__":
    main()
```
